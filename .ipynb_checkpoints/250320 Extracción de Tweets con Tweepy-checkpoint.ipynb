```python
#Obtener informacion de un usuario
data = api.me()
print (json.dumps(data._json, indent=4))
```

```python
#Obtener informacion de otros usuarios
data = api.get_user("nike")
print (json.dumps(data._json, indent=4))
```

```python
#Obtener followers
data = api.followers(screen_name="nike")
```

```python
# Obtener un grupo de followees
# para ello utilizamos un arreglo de tipo user:
for user in data:
    print (json.dumps(user._json, indent=4))
```

```python
# Vemos la longitud del arreglo (nos entrega 20 usuarios, 
# porque la API de tweeter no entrega todos los resultados 
# de una , sino que paginados -en grupos de 20-):
print(len(data))
```

```python
# Cursor es una objeto de tweepy:
# queremos 100 followers de un usuario (nike) -siguen a Nike-
for user in tweepy.Cursor(api.followers, screen_name="nike").items(2):
    print (json.dumps(user._json, indent=4))
```

```python
# followees -personas que Nike sigue-
for user in tweepy.Cursor(api.friends, screen_name="nike").items(2):
    print (json.dumps(user._json, indent=4))
```

```python

#Obtener un timeline de un usuario (todos los tweets que ha hecho un usuario)
print('Obtener un timeline de un usuario todos los tweets que ha hecho un usuario')
# tweet_mode determina la longitud del tweet a obtener:
for tweet in tweepy.Cursor(api.user_timeline, screen_name="nike", tweet_mode="extended").items(2):
    print (json.dumps(tweet._json,  indent=4))
    print (tweet._json['full_text'])
```

```python
#Buscar Tweets
for tweet in tweepy.Cursor(api.search, q="rusia 2018", tweet_mode="extended").items(2):
    print (json.dumps(tweet._json, indent=4))
```
